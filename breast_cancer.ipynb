## Breast Cancer Classification with simple Neural Network (NN)

```python
# importing the dependencies
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection and Processing

```python
# loading the data from sklearn
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
# loading the data to a data frame
data_frame=pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
# print the first 5 rows of the data
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer.target
```

```python
# print the last 5 rows of the data
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the dataset
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures of the data
data_frame.describe()
```

```python
# checking the distribution of the target variable
data_frame['label'].value_counts()
```

```python
data_frame.groupby('label').mean()
```

```python
a=data_frame.drop(columns='label',axis=1)
b=data_frame['label']
```

```python
print(a)
```

```python
print(b)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(a, b, test_size=0.2, random_state=2)
```

```python
print(a.shape,X_train.shape,X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
a_train_std = scalar.fit_transform(X_train)
a_test_std = scalar.transform(X_test)
```

Building the Neural Network

```python
# importing tensorflow and Keras
import tensorflow as tf 
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Neural Network
history = model.fit(a_train_std, Y_train, validation_split=0.1,epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(a_test_std, Y_test)
print('Accuracy:', accuracy)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

# Assuming you have a scaler object created and fitted earlier in your code
scaler = StandardScaler()
 # Make sure to fit the scaler on your training data

input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# Change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

scaler.fit(input_data_reshaped) 
# Standardizing the input data using the previously fitted scaler
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if prediction_label[0] == 0:
    print('The tumor is Malignant')
else:
    print('The tumor is Benign')
```
